```python
from win32com.client import GetObject
import base64, requests
import json
import webbrowser

requests.packages.urllib3.disable_warnings()

class LCU:
    def __init__(self):
        self.protocol = 'https'
        self.host = '127.0.0.1'


    def open_request(self,riotclientToken, riotclientPort, url):
        token = base64.b64encode(('riot:' + riotclientToken).encode("ascii")).decode('ascii')
        header = {'Accept': 'application/json, text/plain, */*', 'Authorization' : f'Basic {token}'}

        u = f'{self.protocol}://{self.host}:{riotclientPort}{url}'
        r = requests.get(u, headers=header, verify=False)
        return r.text

def grab_gamedata() :
    riotclientToken = False
    riotclientPort = False
    winmgmts = GetObject('winmgmts:')
    processes = winmgmts.InstancesOf('Win32_Process')
    for p in processes:
        if p.Properties_("Name").Value == 'LeagueClientUx.exe' :
            gamedata = p.Properties_[1].Value

    for i in gamedata.split(" ") :
        if "--riotclient-auth-token" in i :
            riotclientToken = i.replace("--riotclient-auth-token=","")
            riotclientToken = riotclientToken.replace('"','')
        elif "--riotclient-app-port" in i :
            riotclientPort = i.replace("--riotclient-app-port=","")
            riotclientPort = riotclientPort.replace('"','')
    return riotclientToken, riotclientPort

def find_client():
    riotclientToken = False
    riotclientPort = False
    while riotclientToken == False and riotclientPort == False :
        riotclientToken, riotclientPort = grab_gamedata()
        if riotclientToken != False and riotclientPort != False :
            print("Client found")
    return riotclientToken, riotclientPort

def main(riotclientToken, riotclientPort) :
    lcu = LCU()
    region = json.loads(lcu.open_request(riotclientToken, riotclientPort,"/riotclient/get_region_locale"))["region"]

    players = lcu.open_request(riotclientToken, riotclientPort,'/chat/v5/participants/champ-select')
    list_players = []
    for i in json.loads(players)["participants"] :
        list_players.append(i["name"])

    link = 'https://www.op.gg/multisearch/%s?summoners='%region
    for i in list_players :
        link = link + "%s,"%i

    if len(list_players) == 0 :
        print("Lobby not found")
        return False
    else :
        print("Lobby found")
        webbrowser.open('%s'%link, new=2)
        return True

try :
    riotclientToken, riotclientPort = find_client()
    print("You have to be connected to the chat lobby")
    status = False
    while status == False :
        status = main(riotclientToken, riotclientPort)
        if status == False :
            d=input("Press a key to retry")
except :
    print("Client not found")

k=input("Press a key to exit") 
```

```python
from lcu_driver import Connector
import time

connector = Connector()

async def create_custom_lobby(connection):
  custom = {
    'customGameLobby': {
      'configuration': {
        'gameMode': 'PRACTICETOOL',
        'gameMutator': '',
        'gameServerRegion': '',
        'mapId': 11,
        'mutators': {'id': 1},
        'spectatorPolicy': 'AllAllowed',
        'teamSize': 5
      },
      'lobbyName': 'PRACTICETOOL',
      'lobbyPassword': ''
    },
    'isCustom': True
  }
  await connection.request('POST', '/lol-lobby/v2/lobby', data=custom)

@connector.ready
async def connect(connection):
  await create_custom_lobby(connection)

@connector.close
async def disconnect(connection):
    print('Task finished, closing in : ')
    print("3")
    time.sleep(1)
    print("2")
    time.sleep(1)
    print("1")
    time.sleep(1)
    raise SystemExit("Leaving")

connector.start()
```
